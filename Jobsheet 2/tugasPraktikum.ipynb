```python
import pandas as pd

pd.set_option('display.max_columns', None)
df = pd.read_csv('wbc.csv')
```

```python
df_selected = df.drop(columns=['id', 'Unnamed: 32'])
```

```python
df_selected.info()
```

```python
df_selected.head()
```

```python
from sklearn.preprocessing import LabelEncoder, StandardScaler
```

```python
le = LabelEncoder()
df_selected['diagnosis'] = le.fit_transform(df_selected['diagnosis'])
```

```python
df_selected.head(50)
```

```python
df_selected.info()
```

```python
std = StandardScaler()
numeric_columns = df_selected.select_dtypes(include=['float64']).columns

df_selected[numeric_columns] = std.fit_transform(df_selected[numeric_columns])
```

```python
df_selected.head()
```

```python
from sklearn.model_selection import train_test_split

# Membagi data menjadi 80% data latih dan 20% data uji
df_selected_train, df_selected_test = train_test_split(df_selected, test_size=0.2, random_state=0, stratify=df_selected['diagnosis'])

print(f'Jumlah label data asli:\n{df_selected.diagnosis.value_counts()}')
print(f'Jumlah label data train:\n{df_selected_train.diagnosis.value_counts()}')
print(f'Jumlah label data test:\n{df_selected_test.diagnosis.value_counts()}')
```
